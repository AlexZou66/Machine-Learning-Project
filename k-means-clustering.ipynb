```python
df=pd.read_csv('/kaggle/input/tencent1/11.csv')
df
```

```python
filt = (df['play number'].str.contains('w'))
df['play number'] = df['play number'].str.replace('w','')
df['play number'] = df['play number'].str.replace('y','')
df["play number"] = pd.to_numeric(df["play number"])
df1=df.loc[filt==False, 'play number']*10000
df1 = df1.fillna(value=0)
df2=df.loc[filt==True, 'play number']
df2 = df2.fillna(value=0)
df3=pd.concat([df1, df2]).sort_index()
df3
```

```python
from scipy import stats
t=stats.zscore(df3) 
#print(t)
t=abs(t) 
filtered_entries = t < 0.08
#print(filtered_entries)
new_df = df3[filtered_entries]
new_df
```

```python
normalized_df=(new_df-new_df.mean())/new_df.std()
normalized_df
```

```python
from sklearn.cluster import KMeans
import numpy as np
X = np.array(new_df).reshape(-1,1)
X
```

```python
result=[]
for i in range(10):
    kmeans = KMeans(n_clusters=i+1, random_state=0).fit(X)
    result.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot([i+1 for i in range(10)],result)
#plt.plot(kmeans.cluster_centers_,markersize=500)
```
